```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt

model = keras_cv.models.StableDiffusion(img_width=256, img_height=256)  # Reduce image size

# Lower batch size
batch_size = 2

# Process text to image without visualization
start_time = time.time()
images = model.text_to_image("alan turing drinking coffee", batch_size=batch_size)
end_time = time.time()
processing_time = end_time - start_time

print("Processing time:", processing_time, "seconds")
print("Number of images processed:", batch_size)

# Show the generated images
for image in images:
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
import time
import random
import string
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt

# Function to generate random text prompts
def generate_random_prompt(length=20):
    return ''.join(random.choices(string.ascii_lowercase + ' ', k=length))

# Define model with smaller image size
model = keras_cv.models.StableDiffusion(img_width=256, img_height=256)

# Define batch size
batch_size = 1

# Process random text prompt to generate an image without visualization
random_prompt = generate_random_prompt()
try:
    images = model.text_to_image(random_prompt, batch_size=batch_size)
    for image in images:
        if image is not None:
            plt.imshow(image)
            plt.axis('off')
            plt.show()
        else:
            print("Generated image is None.")
except Exception as e:
    print("Error occurred:", e)

```
